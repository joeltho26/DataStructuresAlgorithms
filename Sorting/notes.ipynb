```python
def subsequence(string,output,size,index=-1,curr=""):
    if index == size:
        return

    if len(curr) > 0:
        output.append(curr)
    
    i = index + 1
    while i < size:
        curr += string[i]
        subsequence(string,output,size,i,curr)
        curr = curr[0:-1]  
        i += 1  
        
def check_same_sequence(output):
    data = set()
    for i in range(len(output)):
        if output[i] in data:
            return True
        else:
            data.add(output[i])
    return False
    
if __name__ == "__main__":
    input = "geeksforgeeks"
    output = []
    subsequence(input,output,len(input))  
    print(output)
    print(check_same_sequence(output))
    
```
