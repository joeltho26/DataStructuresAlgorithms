```python


class Node:
    def __init__(self,data):
        self.data = data
        self.next = next
        
class Stack:
    def __init__(self,size):
        self.size = size
        self.front = None
        self.rear = None
        self.count = 0
        
    def push(self,data):
        if self.is_full():
            print("Stack is full")
            return
        
        node = Node(data)
        if not self.front:
            self.front = self.rear = node
            self.count += 1
            return
        
        self.rear.next = node
        self.rear = node
        self.count +=1
    
    def pop(self):
        if self.is_empty():
            print("Stack is empty")
            return
        
        if self.get_size() == 1:
            self.count -= 1
            value = self.front.data
            self.front = self.rear = None
            return value
        
        itr = self.front
        while itr.next != self.rear:
            itr = itr.next
        value = itr.next.data
        itr.next = None
        self.rear = itr
        self.count -= 1
        return value
            
    def is_empty(self):
        if self.get_size() == 0:
            return True
        return False
    
    def is_full(self):
        if self.get_size() == self.size:
            return True
        return False
    
    def display(self):
        if self.is_empty():
            print("Stack is empty")
            return

        itr = self.front
        lstr = ""
        while itr:
            if itr != self.rear:
                lstr += str(itr.data) + "<--"
            if itr == self.rear:
                lstr += str(itr.data)
                break
            itr = itr.next
        print(lstr)
        
    def peek(self):
        if self.is_empty():
            print("Stack is empty")
            return
        
        return self.rear.data
    
    def get_size(self):
        return self.count
    
if __name__ == "__main__":
    s = Stack(5)
    s.push(5)
    s.push(10)
    s.push(15)
    s.push(20)
    s.push(25)
    s.display()
    print(s.get_size())
    print(s.is_full())
    print(s.is_empty())
    print(s.pop())
    print(s.peek())
    s.display()
        
```
