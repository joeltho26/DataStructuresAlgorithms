```python
class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_end(self,data):
        if self.head is None:
            self.head = Node(data,None)
            return
        
        itr = self.head
        while itr.next:
            itr = itr.next
        itr.next = Node(data,itr.next)
    
    def print_list(self):
        if self.head is None:
            print("Linked list is empty")
            return
        
        itr = self.head
        lstr = ""
        while itr:
            if itr.next:
                lstr += str(itr.data) + "<--"
            if not itr.next:
                lstr += str(itr.data)
            itr = itr.next
        print(lstr)
    
    def delete_at_end(self):
        if self.head is None:
            print("Linked list is empty")
            return
        
        itr = self.head
        while itr.next.next:
            itr = itr.next
        value = itr.next.data    
        itr.next = itr.next.next
        return value
    
    def get_count(self):
        count = 0
        itr = self.head
        while itr:
            count += 1
            itr = itr.next
        return count
    
    def get_last_node(self):
        if self.head is None:
            print("Linked list is empty")
            return
        
        itr = self.head
        while itr.next:
            itr = itr.next
        return itr
    
class Stack:
    def __init__(self,size):
        self.size = size
        self.container = LinkedList()
        
    def push(self,data):
        if self.container.get_count() == self.size:
            print("Stack is full")
            return
        
        self.container.insert_at_end(data)
    
    def pop(self):
        if self.container.get_count() == 0:
            print("Stack is empty!")
            return
        return self.container.delete_at_end()
    
    def peek(self):
        if self.container.get_count() == 0:
            print("Stack is empty!")
            return
        return self.container.get_last_node().data
    
    def get_size(self):
        return self.container.get_count()
    
    def is_empty(self):
        if self.container.get_count() == 0:
            return True
        return False
    
    def is_full(self):
        if self.container.get_count() == self.size:
            return True
        return False
    
    def display(self):
        self.container.print_list()
        
if __name__ == "__main__":
    s = Stack(5)
    s.push(5)
    s.push(10)
    s.push(15)
    s.push(20)
    s.push(25)
    s.display()
    print(s.get_size())
    print(s.is_full())
    print(s.is_empty())
    print(s.pop())
    print(s.peek())
    s.display()
```
