```python
class Stack:
    def __init__(self,size):
        self.size = size
        self.container = []
        
    def push(self,data):
        if self.is_full():
            print("Stack is full")
            return
        
        self.container.append(data)
        self.sort_values()
        self.reverse_array()
        
        #self.container.sort()
        #self.container.reverse()
    
    def sort_values(self):
        size = len(self.container)
        for i in range(size -1):
            swapped = False
            for j in range(size - i - 1):
                if self.container[j] > self.container[j+1]:
                    self.container[j], self.container[j+1] = self.container[j+1], self.container[j]
                    swapped = True
                    
            if not swapped:
                break
            
    def reverse_array(self):
        if self.is_empty():
            print("Stack is empty")
            return
        data_list = []
        for i in range(len(self.container)-1,-1,-1):
            data_list.append(self.container[i])
        self.container = data_list
    
    def pop(self):
        if self.is_empty():
            print("Stack is empty")
            return
        
        return self.container.pop()
    
    def peek(self):
        return self.container[-1]
    
    def get_size(self):
        return len(self.container)
    
    def is_empty(self):
        if len(self.container) == 0:
            return True
        return False
    
    def is_full(self):
        if len(self.container) == self.size:
            return True
        return False
    
    def display(self):
        if self.is_empty():
            print("Stack is empty")
            return
        
        for i in range(len(self.container)):
            if i == len(self.container) - 1:
                print(self.container[i])
            if i != len(self.container) - 1:
                print(self.container[i], end = "<--")

if __name__ == "__main__":
    s = Stack(5)
    s.push(5)
    s.push(15)
    s.push(10)
    s.push(25)
    s.push(20)
    s.display()
    print(s.get_size())
    print(s.is_full())
    print(s.is_empty())
    print(s.pop())
    s.display()
```
