```python
class Node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
        
class Stack:
    def __init__(self,size):
        self.size = size
        self.front = None
        self.rear = None
        self.count = 0
        self.reverse_front = self.reverse_rear = None
        
    def push(self,data):
        if self.is_full():
            print("Stack is full!")
            return
        
        node = Node(data)
        if not self.front:
            self.count += 1
            self.front = self.rear = node
            return
        
        self.rear.next = node
        self.rear = node
        self.count += 1
        self.sort_values()
        self.reverse_values()

    
    def reverse_values(self):
        if self.is_empty():
            print("Stack is empty!")
            return
        
        self.reverse_front = self.reverse_rear = None
        itr = self.front
        while itr:
            self.insert_at_beginning(itr.data)
            if itr.data == self.rear.data:
                break
            itr = itr.next
            
        self.front = self.reverse_front
        self.rear = self.reverse_rear
        
    def insert_at_beginning(self,data):
        if self.is_full():
            print("Stack is full")
            return
        
        if not self.reverse_front:
            node = Node(data,None)
            self.reverse_front = self.reverse_rear = node
            return
        
        node = Node(data,self.reverse_front)
        self.reverse_front = node
        
    def pop(self):
        if self.is_empty():
            print("Stack is empty!")
            return
        
        if self.get_size() == 1:
            self.count -= 1
            value = self.front.data
            self.front = self.rear = None
            return value
        
        itr = self.front
        while itr.next.next:
            itr = itr.next
        value = itr.next.data
        itr.next = None
        self.rear = itr
        self.count -= 1
        return value
    
    def peek(self):
        if self.is_empty():
            print("Stack is empty!")
            return
        
        itr = self.front
        while itr.next:
            itr = itr.next
        return itr.data
    
    def is_empty(self):
        if self.get_size() == 0:
            return True
        return False
    
    def get_size(self):
        count = 0
        itr = self.front
        while itr:
            count += 1
            itr = itr.next
        return count 
    
    def is_full(self):
        if self.get_size() == self.size+1:
            return True
        return False
    
    def sort_values(self):
        if self.is_empty():
            print("Stack is empty!")
            return
        
        swapped = True
        while swapped:
            swapped = False
            itr = self.front
            while itr.next:
                if itr.data > itr.next.data:
                    itr.next.data, itr.data = itr.data, itr.next.data
                    swapped = True
                itr = itr.next
                    
            if not swapped:
                break
            
    def display(self):
        if self.front is None:
            print("Stack is empty!")
            return
        
        itr = self.front
        lstr = ""
        while itr:
            if itr.data == self.rear.data:
                lstr += str(itr.data)
            if itr.data != self.rear.data:
                lstr += str(itr.data) + "<--"
            itr = itr.next
        print(lstr)
        
if __name__ == "__main__":
    s = Stack(5)
    s.push(5)
    s.push(15)
    s.push(10)
    s.push(25)
    s.push(20)
    s.display()
    print(s.get_size())
    print(s.is_full())
    print(s.is_empty())
    print(s.pop())
    print(s.peek())
    s.display()
```
