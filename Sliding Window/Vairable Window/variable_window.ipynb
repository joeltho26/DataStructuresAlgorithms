```python
# longest substring with unique characters

def longest_sub_array_unique(string):
    max = 0
    output = []
    for i in range(len(string)):
        for j in range(i+1,len(string)+1):
            if len(string[i:j]) == len(set(string[i:j])):
                if len(string[i:j]) > max:
                    output = [string[i:j]]
                    max = len(string[i:j])
                elif len(string[i:j]) == max:
                    output.append(string[i:j])
                    
    return output, max

strings = ["abcabcbb","bbbbb","pwwkew"]
for string in strings:
    print(longest_sub_array_unique(string))
```
