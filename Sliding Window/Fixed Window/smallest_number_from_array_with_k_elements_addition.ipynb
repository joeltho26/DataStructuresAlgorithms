```python
from functools import reduce
def sum_of_numbers_k_elements(arr,k):
    min = float('inf')
    res = []
    size = len(arr)
    for i in range(size-k+1):
        sum = reduce(lambda x,y: x+y,arr[i:i+k])
        if sum < min:
            min = sum
        res.append(sum)
    return res, min

arr = [31,69,39,24,86,27,88,56,53,22,93,67]
k = 4
print(sum_of_numbers_k_elements(arr,k))
```
