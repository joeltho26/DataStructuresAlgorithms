```python
from collections import deque
class Queue:
    def __init__(self,size):
        self.size = size
        self.container = deque()
    
    def enqueue(self,data):
        if not self.is_full():
            self.container.appendleft(data)
            self.sort_values()
            self.reverse()
            return
        print("Queue is full")
        return
    
    def dequeue(self):
        if not self.is_empty():
            return self.container.pop()
        print("Queue is empty")
        return
    
    def front(self):
        if not self.is_empty():
            return self.container[-1]
        print("Queue is empty")
        return
    
    def sort_values(self):
        if not self.is_empty():
            size = len(self.container)
            for i in range(size - 1):
                swapped = False
                for j in range(size - i - 1):
                    if self.container[j] > self.container[j + 1]:
                        self.container[j], self.container[j + 1] = self.container[j + 1], self.container[j]
                        swapped = True
                
                if not swapped:
                    break
            return
        print("Queue is empty!")
        return
    
    def reverse(self):
        if not self.is_empty():
            self.container.reverse()
            return
        print("Queue is empty!")
        return
    
    def rear(self):
        if not self.is_empty():
            return self.container[0]
        print("Queue is empty")
        return
    
    def get_size(self):
        return len(self.container)
    
    def is_empty(self):
        if self.get_size() == 0:
            return True
        return False
    
    def is_full(self):
        if self.get_size() == self.size:
            return True
        return False
    
    def display(self):
        for i in range(len(self.container)):
            if i == len(self.container) - 1:
                print(self.container[i])
            if i != len(self.container) - 1:
                print(self.container[i],end="-->")
    
if __name__ == "__main__":
    q = Queue(5)
    q.enqueue(5)
    q.enqueue(25)
    q.enqueue(10)
    q.enqueue(20)
    q.enqueue(15)
    q.display()
    print("Size:",q.get_size())
    print("Is full:",q.is_full())
    print("Is empty:",q.is_empty())
    print("Deleted:",q.dequeue())
    print("Deleted:",q.dequeue())
    print("Front:",q.rear())
    print("Front",q.front())
    q.display()
        
```
