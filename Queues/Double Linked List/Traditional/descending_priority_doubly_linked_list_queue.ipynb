```python
class Node:
    def __init__(self,prev=None,data=None,next=None):
        self.prev = prev
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_beginning(self,data):
        if not self.head:
            self.head = Node(None,data,None)
            return
        
        node = Node(None,data,self.head)
        self.head.prev = node
        self.head = node
        return
    
    def remove_at_end(self):
        if self.head:
            if self.get_length() == 1:
                value = self.head.data
                self.head = None
                return value
            else:
                itr = self.head
                while itr.next.next:
                    itr = itr.next
                value = itr.next.data
                itr.next = None
                return value
        print("Linked list is empty!")
        return
    
    def remove_by_value(self,value):
        if self.head:
            if self.head.data == value:
                self.head = self.head.next
                return
            
            itr = self.head
            while itr.next:
                if itr.next.data == value:
                    itr.next = itr.next.next
                    return
                itr = itr.next
            print(f"Value {value} not found!")
            return                
        print("Linked list is empty!")
        return
    
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            itr = itr.next
            count += 1
        return count
    
    def get_first_node(self):
        if self.head:
            return self.head.data
        print("Linked list is empty!")
        return
    
    def get_last_node(self):
        if self.head:
            itr = self.head
            while itr.next:
                itr = itr.next
            return itr.data
                
        print("Linked list is empty!")
        return
    
    def sort_values(self):
        if self.head:
            swapped = True
            while swapped:
                swapped = False
                itr = self.head
                while itr.next:
                    if itr.data > itr.next.data:
                        itr.data, itr.next.data = itr.next.data, itr.data
                        swapped = True
                    itr = itr.next
                
                if not swapped:
                    return
            return
        print("Linked List is empty!")
        return
    
    def print_nodes(self):
        if self.head:
            size = self.get_length()
            count = 0
            lstr = ''
            itr = self.head
            while itr:
                count += 1
                if count < size:
                    lstr += str(itr.data) + '-->'
                else:
                    lstr += str(itr.data)
                itr = itr.next
            return lstr
        print("Linked list is empty!")
        return 

class Queue:
    def __init__(self,capacity):
        self.capacity = capacity
        self.container = LinkedList()
        
    def get_size(self):
        return self.container.get_length()
    
    def front(self):
        if not self.is_empty():
            print("Front value:",self.container.get_last_node())
            return
        print("Queue is empty!")
        return
            
    def rear(self):
        if not self.is_empty():
            print("Rear value:",self.container.get_first_node())
            return
        print("Queue is empty!")
        return
    
    def enqueue(self,data):
        if not self.is_full():
            self.container.insert_at_beginning(data)
            self.container.sort_values()
            return
        print("Queue is full!")
        return 
    
    def dequeue(self):
        if not self.is_empty():
            print("Deleted:",self.container.remove_at_end())
            return
        print("Queue is empty!")
        return 
    
    def is_empty(self):
        if self.get_size() == 0:
            return True
        return False
    
    def is_full(self):
        if self.get_size() == self.capacity:
            return True
        return False
    
    def display(self):
        if not self.is_empty():
            result = self.container.print_nodes()
            print(result)
            return
        print("Queue is empty!")
        return
    
if __name__ == "__main__":
    q = Queue(5)
    q.enqueue(1)
    q.enqueue(3)
    q.enqueue(5)
    q.enqueue(2)
    q.enqueue(4)
    print("Is queue full:", q.is_full())
    print("Is queue empty:",q.is_empty())
    q.display()
    print("Size:",q.get_size())
    q.rear()
    q.front()
    q.dequeue()
    q.dequeue()
    q.display()
```
