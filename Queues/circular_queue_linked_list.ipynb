```python
#single linked list
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class Queue:
    def __init__(self,size):
        self.size = size
        self.front = self.rear = None
        self.count = 0
    
    def enqueue(self,data):
        if self.is_full():
            print("Queue is full")
            return
        
        node = Node(data)
        if not self.front:
            self.front = self.rear = node
            self.rear.next = self.front
            self.count += 1
            return
        
        self.rear.next = node
        self.rear = node
        self.rear.next = self.front
        self.count += 1
    
    def dequeue(self):
        if self.is_empty():
            print("Queue is empty")
            return
        
        value = self.front.data
        self.front = self.front.next
        self.rear.next = self.front
        self.count -= 1
        
        if not self.front:
            self.rear = None
        return value    
    
    def peek(self):
        if self.is_empty():
            print("Queue is empty") 
            return
        
        itr = self.front
        while itr:
            if itr == self.rear:
                break
            itr = itr.next
        return itr.data
    
    def get_size(self):
        count = 0
        itr = self.front
        while itr:
            count += 1
            if itr == self.rear:
                break
            itr = itr.next
        return count
    
    def is_empty(self):
        if self.get_size() == 0:
            return True
        return False
    
    def is_full(self):
        if self.get_size() == self.size:
            return True
        return False
    
    def display(self):
        if self.is_empty():
            print("Queue is empty")
            return
        
        itr = self.front
        lstr = ""
        while itr:
            if itr != self.rear:
                lstr += str(itr.data) + "<--"
            if itr == self.rear:
                lstr += str(itr.data)
                break
            itr = itr.next
        print(lstr)
    
if __name__ == "__main__":
    q = Queue(5)
    q.enqueue(5)
    q.enqueue(10)
    q.enqueue(15)
    q.enqueue(20)
    q.enqueue(25)
    q.display()
    print(q.get_size())
    print(q.is_full())
    print(q.is_empty())
    print(q.dequeue())
    print(q.peek())
    q.display()
        
```
