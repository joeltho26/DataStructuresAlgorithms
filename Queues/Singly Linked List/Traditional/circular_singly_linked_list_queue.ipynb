```python

class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_beginning(self,data):
        if not self.head:
            self.head = Node(data,self.head)
            self.head.next = self.head
            return
        
        f_node = itr = self.head
        self.head = Node(data,self.head)
        while itr.next != f_node:
            itr = itr.next
        itr.next = self.head
        return
    
    def remove_at_end(self):
        if self.head:
            if self.get_length() == 1:
                value = self.head.data
                self.head = None
                return value
            else:
                f_node = itr = self.head
                while itr.next.next != f_node:
                    itr = itr.next
                value = itr.next.data
                itr.next = itr.next.next
                return value
        print("Linked list is empty!")
        return
    
    def get_length(self):
        count = 0
        f_node = itr = self.head
        while itr:
            if itr.next != f_node:
                count += 1
            else:
                count += 1
                break
            itr = itr.next
        return count
    
    def get_first_node(self):
        if self.head:
            return self.head.data
        print("Linked list is empty!")
        return
    
    def get_last_node(self):
        if self.head:
            f_node = itr = self.head
            while itr.next != f_node:
                itr = itr.next
            return itr.data
                
        print("Linked list is empty!")
        return
    
    def print_nodes(self):
        if self.head:
            lstr = ''
            f_node = itr = self.head
            while itr.next != f_node:
                lstr += str(itr.data) + '-->'
                itr = itr.next
            lstr += str(itr.data)
            return lstr
        print("Linked list is empty!")
        return
    
    def detect_loop(self):
        slow_p = fast_p = self.head
        while fast_p and fast_p.next:
            slow_p = slow_p.next
            fast_p = fast_p.next.next
            if slow_p == fast_p:
                return True
        return False

class Queue:
    def __init__(self,capacity):
        self.capacity = capacity
        self.container = LinkedList()
        
    def get_size(self):
        return self.container.get_length()
    
    def front(self):
        if not self.is_empty():
            print("Front value:",self.container.get_last_node())
            return
        print("Queue is empty!")
        return
            
    def rear(self):
        if not self.is_empty():
            print("Rear value:",self.container.get_first_node())
            return
        print("Queue is empty!")
        return
    
    def enqueue(self,data):
        if not self.is_full():
            self.container.insert_at_beginning(data)
            return
        print("Queue is full!")
        return 
    
    def dequeue(self):
        if not self.is_empty():
            print("Deleted:",self.container.remove_at_end())
            return
        print("Queue is empty!")
        return 
    
    def is_empty(self):
        if self.get_size() == 0:
            return True
        return False
    
    def is_full(self):
        if self.get_size() == self.capacity:
            return True
        return False
    
    def display(self):
        if not self.is_empty():
            result = self.container.print_nodes()
            print(result)
            return
        print("Queue is empty!")
        return
    
    def is_circular(self):
        if self.container.detect_loop():
            print("Is circular:",True)
            return
        print("Is circular:",False)
        return
    
if __name__ == "__main__":
    q = Queue(5)
    q.enqueue(1)
    q.enqueue(3)
    q.enqueue(5)
    q.enqueue(2)
    q.enqueue(4)
    print("Is queue full:", q.is_full())
    print("Is queue empty:",q.is_empty())
    q.display()
    print("Size:",q.get_size())
    q.rear()
    q.front()
    q.dequeue()
    q.dequeue()
    q.display()
    q.is_circular()
```
