```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self,capacity):
        self.capacity = capacity
        self.front = self.rear = None
        self.count = 0
        
    def enqueue(self,data):
        node = Node(data)
        if not self.is_full():
            if self.is_empty():
                self.front = self.rear = node
                self.count += 1
                return
            self.rear.next = node
            self.rear = self.rear.next
            self.count += 1
            return
        print("Queue is full!")
        return 
    
    def dequeue(self):
        if not self.is_empty():
            value = self.front.data
            self.front = self.front.next
            self.count -= 1
            if self.front == None:
                self.rear = None
            print("Deleted:",value)
            return
        print("Queue is empty!")
        return
    
    def display(self):
        if not self.is_empty():
            itr = self.front
            lstr = ''
            while itr.next:
                lstr += str(itr.data) + "<--"
                itr = itr.next
            lstr += str(itr.data)
            print(lstr)
            return
        print("Queue is empty!")
        return        
    
    def is_full(self):
        if self.count == self.capacity:
            return True
        return False
    
    def is_empty(self):
        if self.count == 0:
            return True
        return False
    
    def get_size(self):
        count = 0
        itr = self.front
        while itr:
            itr = itr.next
            count += 1
        return count
    
    def front_value(self):
        if not self.is_empty():
            print("Front value:",self.front.data)
            return
        print("Queue is empty!")
        return
    
    def rear_value(self):
        if not self.is_empty():
            print("Rear value:",self.rear.data)
            return
        print("Queue is empty!")
        return
    
if __name__ == "__main__":
    q = Queue(5)
    q.enqueue(1)
    q.enqueue(2)
    q.enqueue(3)
    q.enqueue(4)
    q.enqueue(5)
    print("Is queue full:", q.is_full())
    print("Is queue empty:",q.is_empty())
    q.display()
    print("Size:",q.get_size())
    q.rear_value()
    q.front_value()
    q.dequeue()
    q.dequeue()
    q.display()
```
