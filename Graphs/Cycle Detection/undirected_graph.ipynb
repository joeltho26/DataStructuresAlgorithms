```python
class Graph:
    def __init__(self, size):
        self.size = size
        self.adj_matrix = [[0] * self.size for _ in range(self.size)]
        self.vertex_data = [""] * self.size
        self.edges = []
    
    def add_edge(self, u, v):
        if 0 <= u < self.size or 0 <= v < self.size:
            self.adj_matrix[u][v] = 1
            self.adj_matrix[v][u] = 1
            
    def add_vertex_data(self, index, node):
        if 0 <= index < self.size:
            self.vertex_data[index] = node
            
    def __dfsUtil(self,vertex_idx, visited, parent):
        visited[vertex_idx] = True
        for i in range(self.size):
            if self.adj_matrix[vertex_idx][i] == 1:
                if not visited[i]:
                    if self.__dfsUtil(i,visited,vertex_idx):
                        return True
                elif parent != i:
                    return True
        return False
    
    def detect_cycle(self):
        visited = [False] * self.size
        for i in range(self.size):
            if not visited[i]:
                if self.__dfsUtil(i,visited,-1):
                    return True
        return False
        
vertices = ['A','B','C','D','E','F','G']
weights = [('D','A'),('A','C'),('A','D'),
           ('A','E'),('E','C'),('C','F'),
           ('C','B'),('C','G'),('B','F')]
    
g = Graph(len(vertices))
for idx, i in enumerate(vertices):
    g.add_vertex_data(idx,i)
    
for i in weights:
    src = g.vertex_data.index(i[0])
    dest = g.vertex_data.index(i[1])
    g.add_edge(src,dest)   
    
print(g.detect_cycle())    
    
```
