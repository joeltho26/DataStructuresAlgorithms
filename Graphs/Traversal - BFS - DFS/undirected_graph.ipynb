```python
class Graph:
    def __init__(self,size):
        self.size = size
        self.adj_matrix = [[0] * self.size for _ in range(self.size)]
        self.vertex_data = [""] * self.size
        
    def add_edge(self, u, v):
        if 0 <= u < self.size and 0 <= v < self.size:
            self.adj_matrix[u][v] = 1
            self.adj_matrix[v][u] = 1
            
    def add_vertex_data(self, index, node):
        if 0 <= index < self.size:
            self.vertex_data[index] = node
            
    def dfs_util(self, vertex_idx, visited, res):
        visited[vertex_idx] = True
        res.append(self.vertex_data[vertex_idx])
        
        for i in range(self.size):
            if self.adj_matrix[vertex_idx][i] == 1 and not visited[i]:
                self.dfs_util(i, visited, res)
            
    def dfs(self, start_vertex):
        res = []
        visited = [False] * self.size
        start_vertex_idx = self.vertex_data.index(start_vertex)
        self.dfs_util(start_vertex_idx, visited, res)
        return res
    
    def bfs(self, start_vertex):
        res = []
        visited = [False] * self.size
        queue = [self.vertex_data.index(start_vertex)]
        visited[queue[0]] = True
        while queue:
            current_item = queue.pop(0)
            res.append(self.vertex_data[current_item])
            
            for v in range(self.size):
                if self.adj_matrix[current_item][v] == 1 and not visited[v]:
                    queue.append(v)
                    visited[v] = True
        return res
        
vertices = ['A','B','C','D','E','F','G']
vertices_data = [('D','A'),('A','C'),('A','D'),
                 ('A','E'),('E','C'),('C','F'),
                 ('C','B'),('C','G'),('B','F')]

g = Graph(len(vertices))

for idx, i in enumerate(vertices):
    g.add_vertex_data(idx,i)

for start,end in vertices_data:
    start_idx = g.vertex_data.index(start)
    end_idx = g.vertex_data.index(end)
    g.add_edge(start_idx,end_idx)
        
print(g.dfs('D'))
print(g.bfs('D'))
```
