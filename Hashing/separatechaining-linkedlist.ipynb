```python
class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            itr = itr .next
            count += 1
        return count
        
    def insert_at_end(self,key,value):
        if self.head is None:
            self.head = Node((key,value),None)
            return
        
        itr = self.head
        while itr.next:
            itr = itr.next
        itr.next = Node((key,value),itr.next)
        
    def search(self,key):
        if not self.head:
            print("Linked list is empty!")
            return
        
        itr = self.head
        while itr:
            if itr.data[0] == key:
                return True
            itr = itr.next
        return False
        
    def find_and_replace(self,key,value):
        if not self.head:
            print("Linked list is empty!")
            return
        
        itr = self.head
        while itr:
            if itr.data[0] == key:
                itr.data[1] = value
            itr = itr.next
        return f"Key: {key} not found in linked list!"
    
    def delete_by_value(self,key):
        if not self.head:
            print("Linked list is empty!")
            return
        
        if self.head.data[0] == key:
            self.head = self.head.next
            return
        
        itr = self.head
        while itr.next:
            if itr.next.data[0] == key:
                value = itr.next.data[1]
                itr.next = itr.next.next
                return value
            itr = itr.next 
        return f"Key: {key} not found in linked list!"
           
    def get_value(self,key):
        if not self.head:
            print("Linked list is empty!")
            return
        
        itr = self.head
        while itr:
            if itr.data[0] == key:
                return itr.data[1]
            itr = itr.next
        return f"Key: {key} not found in linked list!"
    
    def get_first_node(self):
        if not self.head:
            print("Linked list is empty!")
            return
        return self.head.data
    
class Hash:
    def __init__(self,size):
        self.size = size
        self.arr = [None] * self.size
        
    def get_hash(self,key):
        h = 0
        for char in key:
            h += ord(char)
        return h % self.size
        
    def __getitem__(self,key):
        if not self.is_empty():
            h = self.get_hash(key)  
            if self.arr[h] and not isinstance(self.arr[h],LinkedList):
                if self.arr[h][0] == key:
                    return self.arr[h][1]
            elif isinstance(self.arr[h],LinkedList):
                value = self.arr[h].get_value(key)
                return value
            return f"Key: {key} is not found!"
        return "Hash table is empty!"
    
    def __setitem__(self,key,value):
        if not self.is_full():
            h = self.get_hash(key)
            if self.arr[h] and not isinstance(self.arr[h],LinkedList):
                if self.arr[h][0] == key:
                    self.arr[h][1] = value
                    return
                else:
                    data = self.arr[h]
                    self.arr[h] = LinkedList()
                    self.arr[h].insert_at_end(data[0],data[1])
                    self.arr[h].insert_at_end(key,value)
                    return
                    
            elif isinstance(self.arr[h],LinkedList):
                if self.arr[h].search(key):
                    self.arr[h].find_and_replace(key,value)
                    return
                else:
                    self.arr[h].insert_at_end(key,value)
                    return
            else:
                self.arr[h] = (key,value)
                return
        print("Hash table is full!")
        return
    
    def __delitem__(self,key):
        if not self.is_empty():
            h = self.get_hash(key)
            if self.arr[h] and not isinstance(self.arr[h],LinkedList):
                if self.arr[h][0] == key:
                    value = self.arr[h][1]
                    self.arr[h] = None
                    print(value)
                    return
            elif isinstance(self.arr[h],LinkedList):
                value = self.arr[h].delete_by_value(key)
                print(value)
                if self.arr[h].get_length() > 1:
                    return
                else:
                    data = self.arr[h].get_first_node()
                    self.arr[h] = data
                    return
            print(f"Key: {key} is not found!")
            return
        print("Hash table is empty!")
        return
    
    def get_count(self):
        count = 0
        for i in range(len(self.arr)):
            if self.arr[i]:
                count += 1
        return count
                
    
    def is_full(self):
        if self.get_count() == self.size:
            return True
        return False
    
    def is_empty(self):
        if self.get_count() == 0:
            return True
        return False
    
if __name__ == "__main__":
    h = Hash(5)
    h['march 6'] = 10
    h['march 9'] = 20
    h['march 19'] = 30
    h['march 16'] = 40
    h['march 17'] = 50
    print(h.arr)
    print(h.is_empty())
    print(h.is_full())
    del h['march 17']
    del h['march 18']
    print(h['march 19'])
    print(h['march 17'])
    print(h.arr) 
```
