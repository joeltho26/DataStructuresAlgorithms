```python
import math
class Hash:
    def __init__(self,size):
        self.size = size
        self.arr = [None] * self.size
    
    def get_hash(self,key,itr=0):
        h = 0
        for char in key:
            h += ord(char) 
        return (h+itr) % self.size
    
    def rehash(self,up_size=True):
        if up_size:
            prev_arr = self.arr
            prev_size = self.size
            self.increase_size_and_array()
            for i in range(prev_size):
                if prev_arr[i]:
                    k,v = prev_arr[i]
                    self.__setitem__(k,v)
            return
        else:
            prev_arr = self.arr
            prev_size = self.size
            self.decrease_size_and_array()
            for i in range(prev_size):
                if prev_arr[i]:
                    k,v = prev_arr[i]
                    self.__setitem__(k,v)
            return
    
    def increase_size_and_array(self):
        new_size = (2 * self.size) + 1
        self.size = new_size
        self.arr = [None] * self.size
        return
    
    def decrease_size_and_array(self):
        new_size = self.size // 2
        self.size = new_size
        self.arr = [None] * self.size
        return  
        
    def __getitem__(self,key):
        if not self.is_empty():
            h = self.get_hash(key)
            if self.arr[h]:
                if self.arr[h][0] == key:
                    return self.arr[h][1]
                else:
                    for itr in range(1,len(self.arr)):
                        temp_h = self.get_hash(key,itr)
                        if self.arr[temp_h]:
                            if self.arr[temp_h][0] == key:
                                return self.arr[temp_h][1]
            return f"Key: {key} is not found!"
        return "Hash table is empty!"
    
    def __setitem__(self,key,value):
        if not self.is_full():
            h = self.get_hash(key)
            if self.arr[h]:
                if self.arr[h][0] == key:
                    self.arr[h][1] = value
                    return
                else:
                    for itr in range(1,len(self.arr)):
                        temp_h = self.get_hash(key,itr)
                        if self.arr[temp_h]:
                            if self.arr[temp_h][0] == key:
                                self.arr[temp_h][1] = value
                                return
                        else:
                            self.arr[temp_h] = (key,value)
                            return
            else:
                self.arr[h] = (key,value)
                return
        else:
            self.rehash()
            self.__setitem__(key,value)
            return       

    
    def __delitem__(self,key):
        if not self.is_empty():
            if self.get_count() >= math.ceil(self.size/2):
                h = self.get_hash(key) 
                if self.arr[h]:
                    if self.arr[h][0] == key:
                        value = self.arr[h][1]
                        self.arr[h] = None
                        print(value)
                        return
                    else:
                        for itr in range(1,len(self.arr)):
                            temp_h = self.get_hash(key,itr)
                            if self.arr[temp_h]:
                                if self.arr[temp_h][0] == key:
                                    value = self.arr[temp_h][1]
                                    self.arr[temp_h] = None
                                    print(value)
                                    return
                print(f"Key: {key} is not found!")
                return
            else:
                self.rehash(up_size=False)
                self.__delitem__(key)
                return
        print("Hash table is empty!")
        return 
            
    def get_count(self):
        count = 0
        for i in range(len(self.arr)):
            if self.arr[i]: 
                count += 1
        return count      
    
    def is_full(self):
        if self.get_count() == self.size:
            return True
        return False
    
    def is_empty(self):
        if self.get_count == 0:
            return True
        return False
    
if __name__ == "__main__":
    h = Hash(5)
    h['march 6'] = 10
    h['march 9'] = 20
    h['march 19'] = 30
    h['march 16'] = 40
    h['march 17'] = 50
    print(h.arr)
    print(h.is_empty())
    print(h.is_full())
    h['march 1'] = 90
    print(h.arr)
    print(h.get_count())
    del h['march 23']
    del h['march 17']
    print(h.arr)
    del h['march 19']
    print(h.arr)  
    print(h['march 4'])
    print(h['march 29'])
    print(h.arr) 
    print(h.get_count())
```
