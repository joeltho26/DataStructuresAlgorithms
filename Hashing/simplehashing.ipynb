```python

# collision will happen most frequently
class Hash:
    def __init__(self,size=None):
        self.size = size
        self.arr = [None] * size
        
    def get_hash(self,key):
        h = 0
        for char in key:
            h += ord(char)
        return h%self.size
    
    def __getitem__(self,key):
        if not self.is_empty():
            h = self.get_hash(key)
            if self.arr[h]:
                if self.arr[h][0] == key:
                    return self.arr[h][1]
            return f"Key: {key} not found!"
        else:
            return "Hash table is empty!"
        
    
    def __setitem__(self,key,value):
        if not self.is_full():
            h = self.get_hash(key)
            if self.arr[h]:
                if self.arr[h][0] == key:
                    self.arr[h][1] = value
                    return
                else:
                    self.arr[h] = (key,value)
                    return
            else:
                self.arr[h] = (key,value)
                return
        else:
            print("Hash table is full")
            return
    
    def __delitem__(self,key):
        if not self.is_empty():
            h = self.get_hash(key)
            if self.arr[h]:
                if self.arr[h][0] == key:
                    value = self.arr[h][1]
                    self.arr[h] = None
                    print(value)
                    return
            print(f"Key: {key} not found!")
            return
        print("Hash table is empty!")
        return
    
    def get_count(self):
        count = 0
        for i in range(len(self.arr)):
            if self.arr[i]:
                count += 1
        return count
    
    def is_empty(self):
        if self.get_count() == 0:
            return True
        return False
    
    def is_full(self):
        if self.get_count() == self.size:
            return True
        return False   
    
if __name__ == "__main__":
    h = Hash(10)
    h['march 9'] = 90
    h['march 17'] = 30
    print(h.arr)
    h['march 6'] = 40
    h['march 19'] = 10
    print(h.arr)
    print(h["march 17"])
    print(h["march 19"])
    del h["march 17"]
    del h["march 9"]
    print(h.arr)
    print(h.is_empty())
    print(h.is_full())
    print(h.get_count())
```
