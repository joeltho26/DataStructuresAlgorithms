```python
class Hash:
    def __init__(self,size):
        self.size = size
        self.arr = [[] for i in range(self.size)]
        
    def get_hash(self,key):
        h = 0
        for char in key:
            h += ord(char)
        return h%self.size
    
    def __getitem__(self,key):
        if not self.is_empty():
            h = self.get_hash(key)
            for ele in self.arr[h]:
                if ele[0] == key:
                    return ele[1]
            return f"Key: {key} not found!"
        return "Hash table is empty!"
        
    
    def __setitem__(self,key,val):
        if not self.is_full():
            h = self.get_hash(key)
            is_found = False
            for idx,ele in enumerate(self.arr[h]):
                if len(ele) == 2 and ele[0] == key:
                    self.arr[h][idx] = (key,val)
                    is_found = True
                    return
                
            if not is_found:
                self.arr[h].append((key,val)) 
                return
            
        print("Hash table is full!")
        return
            
                
    def __delitem__(self,key):
        if not self.is_empty():
            h = self.get_hash(key)
            for idx,ele in enumerate(self.arr[h]):
                if ele[0] == key:
                    value = self.arr[h][idx][1]
                    del self.arr[h][idx]
                    print(value)
                    return
            print(f"Key: {key} not found!")
            return
        print("Hash table is empty!")
        return
    
    def get_count(self):
        count = 0
        for i in range(self.size):
            if self.arr[i]:
                count += 1
        return count
    
    def is_empty(self):
        if self.get_count() == 0:
            return True
        else:
            return False
    
    def is_full(self):
        if [] in self.arr:
            return False
        return True
    
    def get_total_no_items(self):
        total_items = 0
        for i in range(self.size):
            if self.arr[i]:
                total_items += len(self.arr[i])
        return total_items
        
if __name__ == "__main__":
    h = Hash(10)
    h['march 9'] = 90
    h['march 17'] = 30
    print(h.arr)
    h['march 6'] = 40
    h['march 19'] = 10
    print(h.arr)
    print(h["march 17"])
    print(h["march 18"])
    del h["march 17"]
    del h["march 27"]
    print(h.arr)
    print(h.get_count())
    print(h.is_empty())
    print(h.is_full())
    print(h.get_total_no_items())
    
```
