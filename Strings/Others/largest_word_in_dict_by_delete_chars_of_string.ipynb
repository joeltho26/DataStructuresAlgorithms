```python
# find largest word in dictionary by deleting some characters of given string

def longest(str_data,str):
    max = 0
    res = []
    str_data.sort(key=lambda x: len(x), reverse=True)
    for data in str_data:
        count = 0
        for i in data:
            if i in str:
                count += 1
                index = str.index(i)
                str = str[:index] + str[index+1:]
            else:
                break
        if count == len(data) and len(data) > max:
            res.append(data)
                
    return res
        

str_data = ["ale", "apple", "monkey", "plea"]
str = "abpcplea"
print(longest(str_data,str))

str_data = [ "pintu", "geeksfor", "geeksgeeks", " forgeek" ]
str ="geeksforgeeks"
print(longest(str_data,str))
```
