```python
#check a string is a panagram
import re

def panagram(string,items):
    items = set()
    for i in range(len(string)):
        if string[i] not in items:
            items.add(string[i])
        if len(items) == 26:
            return True
    return False
    
items = set()
string = ["The quick brown fox jumps over the lazy dog","The quick brown fox jumps over the dog"]
for i in string:
    i = re.sub("[^a-z]+","",i.lower())
    print(panagram(i,items))
```
