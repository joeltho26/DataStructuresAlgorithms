```python
# permutation of a string

def permutation_string(index,string,output):
    if index == len(string):
        output.append("".join(string))
        
    for i in range(index,len(string)):
        string[index], string[i] = string[i], string[index]
        permutation_string(index+1,string,output)
        string[index], string[i] = string[i], string[index]

string = "geeks"
output = []
permutation_string(0,list(string),output)
print(len(output))
```
