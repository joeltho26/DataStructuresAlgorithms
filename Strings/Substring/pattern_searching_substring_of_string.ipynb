```python
# search for a matching substring and return the start indexes

def naive_algo(string,pattern):
    matches = []
    itr = len(string)
    count = 0
    for i in range(itr):
        try:
            index = string.index(pattern,count)
            if index not in matches:
                matches.append(index)
            count += 1
        except:
            break
    return matches

def naive_algo2(string,pattern):
    size = len(string)
    index_list = []
    pattern_size = len(pattern)
    for i in range(size-pattern_size+1):
        if string[i:i+pattern_size] == pattern:
            index_list.append(i)
    return index_list

str1 = "THIS IS A TEST TEXT"
pattern = "TEST"
print(naive_algo(str1,pattern))

str1 = "AABAACAADAABAABA"
pattern = "AABA"
print(naive_algo(str1,pattern))

str1 = "THIS IS A TEST TEXT"
pattern = "TEST"
print(naive_algo2(str1,pattern))

str1 = "AABAACAADAABAABA"
pattern = "AABA"
print(naive_algo2(str1,pattern))
```
