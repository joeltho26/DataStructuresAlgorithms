```python
# print the all the substrings in a string

def all_substrings_in_a_string(string):
    substrings = []
    count = 0
    for i in range(len(string)):
        for j in range(i+1,len(string)+1):
            substrings.append(string[i:j])
    return substrings

b = "aaaab"
print(all_substrings_in_a_string(b))
```
