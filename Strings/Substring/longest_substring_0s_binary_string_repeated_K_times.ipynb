```python
# longest substring of 0's from a binary string which is repeated K times

def longest_substring_binary_string_K_times(string,k):
    max = 0
    string = string * k
    size = len(string) 
    for i in range(size):
        for j in range(i+1,size+1):
            if int(string[i:j]) == 0:
                if len(string[i:j]) > max:
                    output = [string[i:j]]
                    max = len(string[i:j])
                elif len(string[i:j]) == max:
                    output.append(string[i:j])
    return output, max

a = "100001"
k = 3
print(longest_substring_binary_string_K_times(a,k))

a = "0100010000"
k = 4
print(longest_substring_binary_string_K_times(a,k))
```
