```python
# longest substring with k unique characters in a given string
def longestKSubstr(string,k):
    max = 0
    size = len(string)
    res = []
    for i in range(size):
        for j in range(i+1,size+1):
            if len(set(string[i:j])) == 3:
                if len(string[i:j]) > max:
                    res = [string[i:j]]
                    max = len(string[i:j])
                elif len(string[i:j]) == max:
                    res.append([string[i:j]])
    return res, max

a = "aabacbebebe"
k = 3
print(longestKSubstr(a,k))
```
