```python
# count of K-size substrings having palindromic permutations

def substring(string,k):
    res = []
    size = len(string)
    output = []
    for i in range(size):
        for j in range(i+1,size+1):
            if len(string[i:j]) == k:
                res.append(string[i:j])

    count = 0
    for i in range(len(res)):
        if reverse(res[i]) in res:
            output.append(res[i])
            count += 1
        
    return output, count
            
def reverse(str2):
    return str2[::-1]
    
a = "abbaca"
k = 3
print(substring(a,k))

a = "aaaa"
k = 1
print(substring(a,k))
 
```
