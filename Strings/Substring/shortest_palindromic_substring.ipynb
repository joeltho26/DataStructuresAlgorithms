```python
# shortest palindrome substring based on length and the precedence of alphabetic order if substring of same length

def substring(string):
    min = float('inf')
    size = len(string)
    for i in range(size):
        for j in range(i+1,size+1):
            if is_palindromic(string[i:j]):
                if len(string[i:j]) < min:
                    output = string[i:j]
                    min = len(string[i:j])
                elif len(string[i:j]) == min:
                    if string[i:j] < output:
                        output = string[i:j]
    return output

def is_palindromic(string):
    return string == string[::-1]

a = "zyzz"
print(substring(a))

a = "abab"
print(substring(a))
```
