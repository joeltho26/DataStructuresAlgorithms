```python
def lengthOfLongestSubstring(s: str):
    if len(s) < 1:
        return 0 

    max = 0
    for i in range(len(s)):
        for j in range(len(s),i,-1):
            val = set(s[i:j])
            if len(val) == len(s[i:j]) and len(val) > max:
                max = len(val)
                
            if len(s[i:j]) <= max+1:
                break
    return max

string = "geeks"
print(lengthOfLongestSubstring(string))
```
