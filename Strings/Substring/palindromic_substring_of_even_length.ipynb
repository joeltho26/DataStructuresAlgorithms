```python
# print the set of palindromic substrings of even length

def check_palindrome_substring_even_length(string):
    size = len(string)
    res = []
    for i in range(size):
        for j in range(i+1,size+1):
            if is_palindrome(string[i:j]) and len(string[i:j])%2 == 0:
                res.append(string[i:j])
    return res
            
            
def is_palindrome(string):
    return string == string[::-1]

a = "aassss"
print(check_palindrome_substring_even_length(a))

a = "gfg"
print(check_palindrome_substring_even_length(a))
```
