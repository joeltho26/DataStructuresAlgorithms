```python
#find the count of palindromic sub-string of a string in its sorted form

def substring(string):
    sub_strings = []
    for i in range(len(string)):
        for j in range(i+1,len(string)+1):
            sub_strings.append(string[i:j])
    return sub_strings

def is_palindrome(substring):
    return substring == substring[::-1]

def count_palindromic_substring(string):
    res = substring(string)
    print(res)
    count = 0
    output = []
    for i in res:
        if is_palindrome(i):
            output.append(i)
            count += 1
    return output,count
    
a = "acbbd"
b = "abbabdbd"
print(count_palindromic_substring(a))

print(count_palindromic_substring(b))
```
