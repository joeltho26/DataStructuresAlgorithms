```python
# print the occurences of substrings in a string the same as the other string

def print_substrings(S1,S2):
    substrings = []
    count = 0
    for i in range(len(S2)):
        for j in range(i+1,len(S2)+1):
            if S1 == S2[i:j]:
                substrings.append((S2[i:j],S2.find(S1,count)))
                count += 1 
    return substrings

S2 = "GeeksforGeeks"
S1= "Geeks"
print(print_substrings(S1,S2),end="\n")

S1 = "GFG"
S2= "g"
print(print_substrings(S1,S2))
```
