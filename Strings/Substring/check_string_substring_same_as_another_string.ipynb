```python
# a string exist as substring in another string and return its matched index

def substring(S1,S2):
    substrings = []
    for i in range(len(S2)):
        for j in range(i+1,len(S2)+1):
            substrings.append(S2[i:j])
    
    for i in substrings:
        if S1 == i:
            return True
    return False

def substring_with_index(S1,S2):
    substrings = []
    for i in range(len(S2)):
        for j in range(i+1,len(S2)+1):
            if S1 == S2[i:j]:
                return S2.find(S1)
    return -1


S1 = "for"
S2= "geeks"
print(substring(S1,S2))
print(substring_with_index(S1,S2))

S1 = "for"
S2= "geeksforgeeks"
print(substring(S1,S2))
print(substring_with_index(S1,S2))
```
