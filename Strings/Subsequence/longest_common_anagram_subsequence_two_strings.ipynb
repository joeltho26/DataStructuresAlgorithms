```python
# largest common anagram subsequences from two different strings

def subsequences(string,size,output,index=-1,curr=""):
    if index == size:
        return
    
    if len(curr) > 0:
        output.append(curr)
        
    i = index + 1
    while i < size:
        curr +=  string[i]
        subsequences(string,size,output,i,curr)
        curr = curr[0:-1]
        i += 1 

def long_common_anagram(str1,str2):
    output1 = []
    subsequences(str1,len(str1),output1)
    output1.sort(key=lambda x: len(x), reverse=True)
    
    output2 = []
    subsequences(str2,len(str2),output2)
    output2.sort(key=lambda x: len(x), reverse=True)
    
    output = None
    max = 0
    for i in range(len(output1)):
        for j in range(len(output2)):
            if len(output1[i]) == len(output2[j]):
                count = 0
                itr = len(output1[i])
                compare_to = output2[j]
                while itr:
                    if output1[i][count] in compare_to:
                        compare_to = compare_to.replace(output1[i][count],"",1)
                        count += 1
                        itr = itr - 1
                    else: 
                        break

                    if len(compare_to) == 0:
                        if len(output1[i]) > max:
                            output = [(output1[i],output2[j])]
                            max = len(output1[i])
                        elif len(output1[i]) == max:
                            output.append((output1[i],output2[j]))
                    
                    
    return output

str1 = "abdacp"
str2 = "ckamb"
print(long_common_anagram(str1, str2))

str1 = "abbcfke"
str2 = "fbaafbly"
print(long_common_anagram(str1, str2))
```
