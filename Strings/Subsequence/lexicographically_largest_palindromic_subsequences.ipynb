```python
# find the lexicographically largest palindromic subsequence of a string

def subsequence(input,size,output,index=-1,curr=""):
    if index==size:
        return
    
    if len(curr) > 0:
        output.append(curr)
        
    i = index + 1
    while i < size:
        curr += input[i]
        subsequence(input, size, output, i, curr)
        curr = curr[0:-1]
        i += 1
        
def is_palindrome(string):
    return string == string[::-1]

def lexicographically_largest_palindrome_subsequence(string):
    output = []
    subsequence(string,len(string),output)
    output.sort(key=lambda x: len(x), reverse=True)
        
    max = ""
    res = None
    for i in output:
        if is_palindrome(i):
            count = 0
            for itr in range(len(i)-1):
                if i[count] <= i[count +1]:
                    count += 1
                else:
                    break

            if count == len(i)-1:
                if i > max:
                    res = [i]
                    max = i
                elif i == max:
                    res.append(i)
                
    return res

input = "abrakadabra"
print(lexicographically_largest_palindrome_subsequence(input))
```
