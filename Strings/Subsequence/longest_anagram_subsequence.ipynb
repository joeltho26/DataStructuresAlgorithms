```python
# longest anagram subsequence

def subsequence(string,size,output,index=-1,curr=""):
    if index == size:
        return
    
    if len(curr) > 0:
        output.append(curr)
        
    i = index + 1
    while i < size:
        curr += string[i]
        subsequence(string,size,output,i,curr)
        curr = curr[0:-1]
        i += 1
    
def longest_anagram(subsequences):
    max = 0
    output = None
    for i in range(len(subsequences)-1):
        for j in range(i+1,len(subsequences)):
            count = 0
            if len(subsequences[i]) == len(subsequences[j]):
                itr = len(subsequences[i])
                compare_from, compare_to = subsequences[i], subsequences[j]
                while itr:
                    if compare_from[count] in compare_to:
                        compare_to = compare_to.replace(subsequences[i][count],"",1)
                        count += 1
                        itr = itr - 1
                    else:
                        break
                    
                    if len(compare_to) == 0:
                        if len(subsequences[i]) > max:
                            output = [(subsequences[i],subsequences[j])]
                            max = len(subsequences[i])
                        elif len(subsequences[i]) == max:
                            output.append((subsequences[i],subsequences[j]))
            
    return output             
                        
string = "geeks"
output = []
subsequence(string,len(string),output)
output.sort(key=lambda x: len(x),reverse=True)
print(longest_anagram(output))
    
    
```
