```python
# minimum number of palindromic subsequences to be removed to empty a binary string
def is_palindrome(string):
    rev = string[::-1]
    if rev == string:
        return True
    return False

def min_removals(string):
    if string == "":
        return 0
    
    if is_palindrome(string):
        return 1
    return 2
    
values = ['010010','01110101011']
for val in values:
    print(min_removals(val))
```
