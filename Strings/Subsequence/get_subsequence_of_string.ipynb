```python
# get the subsequence of a string
def subsequence(string,size,output,index=-1,curr=""):
    if index == size:
        return
    
    if len(curr) > 0:
        output.append(curr)
        
    i = index + 1
    while i < size:
        curr += string[i]
        subsequence(string,size,output,i,curr)
        curr = curr[0:-1]
        i += 1        
        
string = "geeks" 
output = []
subsequence(string,len(string),output)
print(output)
```
