```python
def access_at(string,index,character_count=None):
    if not character_count:
        return string[index]
    
    if character_count:
        return string[index:index+character_count]

if __name__ == "__main__":
    a = "abcbaabcaababcad"
    index = 6
    print(access_at(a,index))
    character_count = 3
    print(access_at(a,index,character_count))
```
