```python
# last element
    
def access_last(string,character_count=None):
    if not character_count:
        return string[-1]

    if character_count:
        return string[-character_count:]
    
a = "abcbaabcaababcad"
print(access_last(a))
character_count = 3
print(access_last(a,character_count))
```
