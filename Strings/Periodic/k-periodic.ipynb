```python

# check a given string is K periodic
def k_periodic(string,k):
    size = len(string)
    value = []
    if size % k == 0:
        for i in range(0,size,k):
            if not value:
                value.append(string[i:i+k])
            if string[i:i+k] in value:
                continue
            else:
                break
        return True 
    return False    

def k_periodic_2(a,k):
    size = len(a)
    for i in range(0,size-k,k):
        if a[i:i+k] == a[i+k:i+(k*k)]:
            continue
        else:
            print(f"String {a} is not {k}-periodic")
            return
    string = a[0:k]
    print(f"String {a} is {k}-periodic: {string}")

string = "geeksgeeks"
k = 5
print(k_periodic(string,k))

a = "geeksforgeeks"
k=3
print(k_periodic(a,k))
```
