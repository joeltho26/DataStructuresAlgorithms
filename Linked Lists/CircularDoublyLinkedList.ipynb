```python
#Circular Doubly Linked List
class Node:
    def __init__(self,prev=None,data=None,next=None):
        self.prev = prev
        self.data = data
        self.next = next
        
class LinkedList:
    def __init__(self):
        self.head = None
     
    def insert_at_beginning(self,data):
        if self.head is None:
            node = Node(None,data,None)
            self.head = node
            self.head.prev = self.head.next = self.head
            return
            
        f_node = itr = self.head
        node = Node(None,data,self.head)
        self.head.prev = node
        self.head = node
        while itr:
            if itr.next == f_node:
                itr.next = node
                self.head.prev = itr
                break
            itr = itr.next
        
    def get_length(self):
        count = 0
        f_node = itr = self.head
        while itr:
            count += 1
            if itr.next == f_node:
                break
            itr = itr.next
        return count
    
    def get_last_node(self):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        f_node = itr = self.head
        while itr:
            if itr.next == f_node:
                return itr
            itr = itr.next
    
    def insert_at_end(self,data):
        if self.head is None:
            self.insert_at_beginning(data)
            return
        
        f_node = itr = self.head
        while itr:
            if itr.next == f_node:
                itr.next = Node(itr,data,self.head)
                self.head.prev = itr.next
                break
            itr = itr.next
    
    def print_forward(self):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        f_node = itr = self.head
        fstr = ""
        while itr:
            fstr += str(itr.data) + "-->"
            if itr.next == f_node:
                break
            itr = itr.next
        print(fstr)
    
    def print_backward(self):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        l_node = itr = self.get_last_node()
        bstr = ""
        while itr:
            bstr += str(itr.data) + "-->"
            if itr.prev == l_node:
                break
            itr = itr.prev
        print(bstr)
        
    def delete_at_end(self):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        f_node = itr = self.head
        while itr.next:
            if itr.next.next == f_node:
                itr.next = self.head
                self.head.prev = itr
                break
            itr = itr.next
    
    def delete_at_beginning(self):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        f_node = itr = self.head
        self.head = self.head.next
        while itr:
            if itr.next == f_node:
                itr.next = self.head
                self.head.prev = itr
                break
            itr = itr.next
    
    def insert_values(self,data_list):
        for data in data_list:
            self.insert_at_end(data)
            
    def insert_at(self,index,data):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        if index < 0 or index > self.get_length():
            raise Exception("Invalid Index")
        
        if index == 0:
            self.insert_at_beginning(data)
            return
        
        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                node = Node(itr,data,itr.next)
                if node.next:
                    itr.next.prev = node
                itr.next = node
                break
            count += 1
            itr = itr.next
            
    def delete_at(self,index):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        if index < 0 or index >= self.get_length():
            raise Exception("Invalid Index")
        
        if index == 0:
            self.delete_at_beginning()
            return
        
        f_node = itr = self.head
        count = 0
        while itr:
            if count == index - 1:
                itr.next = itr.next.next
                if itr.next != f_node:
                    itr.next.prev = itr
                elif itr.next == f_node:
                    self.head.prev = itr
                break
            itr = itr.next
            count += 1
    
    def insert_after_value(self,value,data):
        if self.head is None:
            print("The Linekd list is empty!")
            return
        
        f_node = itr = self.head
        while itr:
            if itr.data == value:
                node = Node(itr,data,itr.next)
                if itr.next == f_node:
                    self.head.prev = node
                elif itr.next != f_node:
                    itr.next.prev = node
                itr.next = node
                return
            if itr.next == f_node:
                break
            itr = itr.next
        print("Value not found!")
        return

    def insert_before_value(self,value,data):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        if self.head.data == value:
            self.insert_at_beginning(data)
            return
        
        f_node = itr = self.head
        while itr:
            if itr.next.data == value:
                node = Node(itr,data,itr.next)
                itr.next.prev = node
                itr.next = node
                return
            if itr.next == f_node:
                break
            itr = itr.next
        print("Value not found!")
        return
    
    def delete_by_value(self,value):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        if self.head.data == value:
            f_node = itr = self.head
            self.head = self.head.next
            while itr:
                if itr.next == f_node:
                    itr.next = self.head
                    self.head.prev = itr
                    return
                itr = itr.next
            
        f_node = itr = self.head
        while itr:
            if itr.next.data == value:
                itr.next = itr.next.next
                if itr.next == f_node:
                    self.head.prev = itr
                elif itr.next != f_node:    
                    itr.next.prev = itr
                return
            if itr.next == f_node:
                break
            itr = itr.next
        print("Value not found!")
        return    
    
    def reverse(self):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        count = 0
        itr = self.head
        size = self.get_length()
        self.head = None
        while itr:
            count += 1
            self.insert_at_beginning(itr.data)
            if count == size:
                break
            itr = itr.next
    
    def sort_values(self):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        swapped = True
        while swapped:
            swapped = False
            f_node = itr = self.head
            while itr and itr.next != f_node:
                if itr.data > itr.next.data:
                    tmp = itr.next.data
                    itr.next.data = itr.data
                    itr.data = tmp
                    swapped = True
                itr = itr.next
            
            if not swapped:
                break     
    
    def search(self,value):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        f_node = itr = self.head
        while itr:
            if itr.data == value:
                return True
            if itr.next == f_node:
                break
            itr = itr.next
        return False
    
    def detect_loop(self):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        slow_p = fast_p = self.head
        while fast_p and fast_p.next:
            slow_p = slow_p.next
            fast_p = fast_p.next.next
            while slow_p == fast_p:
                #self.remove_loop(slow_p)
                return True
        return False
    
    def detect_loop_floyd_cycle(self):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        container = set()
        f_node = itr = self.head
        while itr:
            if itr in container:
                #self.remove_loop(slow_p)
                return True
            container.add(itr)
            if itr.next == f_node:
                break
        return False
    
    def remove_loop(self,pointer):
        ptr1 = ptr2 = pointer
        k = 1
        
        while ptr1 != ptr2.next:
            ptr2 = ptr2.next
            k += 1
            
        ptr1 = ptr2 = self.head    
        for i in range(k):
            ptr2 = ptr2.next
            
        while ptr1 != ptr2:
            ptr1 = ptr1.next
            ptr2 = ptr2.next
        
        while ptr1 != ptr2.next:
            ptr2 =  ptr2.next
            
        ptr2.next = None
          
if __name__ == "__main__":
    ll = LinkedList()
    ll.insert_at_beginning(50)
    ll.insert_at_beginning(40)
    ll.insert_at_beginning(30)
    ll.print_forward()
    ll.print_backward()
    ll.insert_at_end(60)
    ll.insert_at_end(70)
    ll.print_forward()
    ll.print_backward()
    ll.insert_values([80,90])
    ll.print_forward()
    ll.print_backward()
    print(ll.get_length())
    ll.insert_at(3, 55)
    ll.insert_at(0, 20)
    ll.insert_at(9, 100)
    ll.print_forward()
    ll.print_backward()
    ll.delete_at(0)
    ll.delete_at(8)
    ll.delete_at(4)
    ll.print_forward()
    ll.print_backward()
    ll.insert_after_value(5,2)
    ll.insert_after_value(30,35)
    ll.insert_after_value(55,60)
    ll.insert_after_value(90,100)
    ll.print_forward()
    ll.print_backward()
    ll.insert_before_value(5,7)
    ll.insert_before_value(30,20)
    ll.insert_before_value(100,95)
    ll.insert_before_value(40,45)
    ll.print_forward()
    ll.print_backward()
    ll.delete_by_value(5)
    ll.delete_by_value(100)
    ll.delete_by_value(20)
    ll.delete_by_value(55)
    ll.print_forward()
    ll.print_backward()
    ll.delete_at_end()
    ll.delete_at_beginning()
    ll.print_forward()
    ll.print_backward()
    ll.reverse()
    ll.print_forward()
    ll.print_backward()
    ll.sort_values()
    ll.print_forward()
    ll.print_backward()
    print(ll.search(100))
    print(ll.detect_loop())
    print(ll.detect_loop_floyd_cycle())
```
