```python

#Singly Linked Circular List:

class Node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_beginning(self, data):
        if self.head is None:
            self.head = Node(data)
            self.head.next = self.head
            return
        
        fnode = itr = self.head
        self.head = Node(data, self.head)
        while itr.next != fnode:
            itr = itr.next
        itr.next = self.head
    
    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data)
            self.head.next = self.head
            return
        
        fnode = itr = self.head
        while itr.next != fnode:
            itr = itr.next
        itr.next = Node(data, itr.next)
    
    def delete_at_end(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        if self.get_length() == 1:
            self.head = None
            return
        
        fnode = itr = self.head 
        while itr.next.next != fnode:
            itr = itr.next
        itr.next = fnode
    
    def delete_at_beginning(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        if self.get_length() == 1:
            self.head = None
            return
        
        fnode = itr = self.head
        self.head = self.head.next
        while itr.next != fnode:
            itr = itr.next
        itr.next = self.head
    
    def print_forward(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        fstr = ""
        fnode = itr = self.head
        while itr.next != fnode:
            fstr += str(itr.data) + "-->"
            itr = itr.next
        fstr += str(itr.data)
        print(fstr)
    
    def print_backward(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        bstr = ""
        fnode = itr = self.head
        while itr.next != fnode:
            bstr = "<--" + str(itr.data) + bstr
            itr = itr.next
        bstr = str(itr.data) + bstr
        print(bstr)
    
    def get_length(self):
        count = 1
        fnode = itr = self.head
        while itr.next != fnode:
            itr = itr.next
            count += 1
        return count
    
    def insert_at(self, index, data):
        if index < 0 or index > self.get_length():
            print("Invalid index!")
            return
        
        if index == 0:
            self.insert_at_beginning(data)
            return
        
        if index == self.get_length():
            self.insert_at_end(data)
            return 
        
        count = 0
        fnode = itr = self.head
        while itr.next != fnode:
            if count == index - 1:
                itr.next = Node(data, itr.next)
                return
            itr = itr.next
            count += 1
    
    def delete_at(self, index):
        if self.head is None:
            print("Linked list is empty!")
            return False 
        
        if index < 0 or index >= self.get_length():
            print("Invalid index!")
            return
        
        if index == 0:
            self.delete_at_beginning()
            return
        
        fnode = itr = self.head
        count = 0
        while itr.next != fnode:
            if count == index - 1:
                itr.next = itr.next.next
                return
            itr = itr.next
            count += 1
    
    def search(self, val):
        if self.head is None:
            print("Linked list is empty!")
            return False
        
        if self.head.data == val:
            return True
        
        fnode = itr = self.head
        while itr.next != fnode:
            if itr.next.data == val:
                return True
            itr = itr.next   
            
        print("No value found!")
        return False 
    
    def insert_values(self, data_list):
        for data in data_list:
            self.insert_at_end(data)
            
    def insert_after_value(self, val, data):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        count = 0
        itr = self.head
        while count < self.get_length():
            if itr.data == val:
                itr.next = Node(data, itr.next)
                return
            itr = itr.next
            count += 1
            
        print(f"No value found to insert after {val}!")
    
    def insert_before_value(self, val, data):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        if self.head.data == val:
            self.insert_at_beginning(data)
            return
        
        fnode = itr = self.head
        while itr.next != fnode:
            if itr.next.data == val:
                itr.next = Node(data, itr.next)
                return
            itr = itr.next
            
        print(f"No value found to insert before {val}!")
    
    def delete_by_value(self, val):
        if self.head is None:
            print("Linked list is empty!")
            return
                
        if self.head.data == val:
            self.delete_at_beginning()
            return
            
        fnode = itr = self.head
        while itr.next != fnode:
            if itr.next.data == val:
                itr.next = itr.next.next
                return
            itr = itr.next
            
        print(f"No value found to delete {val}!")
    
    def delete_after_value(self, val):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        count = 0
        fnode = itr = self.head
        while count < self.get_length():
            if itr.data == val:
                if itr.next == fnode:
                    # print(f"No node available to delete after value {val}")
                    self.head = self.head.next
                itr.next = itr.next.next
                return
            itr = itr.next
            count += 1
            
        print(f"No value found to delete after value {val}!")
    
    def delete_before_value(self, val):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        if self.head.data == val:
            # print(f"No node available to delete before value {val} at HEAD")
            self.delete_at_end()
            return
        
        if self.head.next.data == val:
            self.delete_at_beginning()
            return
        
        fnode = itr = self.head
        while itr.next.next != fnode:
            if itr.next.next.data == val:
                itr.next = itr.next.next
                return
            itr = itr.next
            
        print(f"No value found to delete before value {val}!")
    
    def sort(self):
        if self.head is None:
            print("Linked list is empty!")
            return 
        
        swapped = True
        while swapped:
            fnode = itr = self.head
            swapped = False
            while itr.next != fnode:
                if itr.data > itr.next.data:
                    itr.data, itr.next.data = itr.next.data, itr.data
                    swapped = True
                itr = itr.next
                
            if not swapped:
                break
    
    def reverse(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        fnode = itr = self.head
        self.head = None
        while itr.next != fnode:
            self.insert_at_beginning(itr.data)
            itr = itr.next
        self.insert_at_beginning(itr.data)
    
    def detect_cycle(self):
        if self.head is None:
            print("Linked list is empty!")
            return False
        
        count = 0
        container = set()
        itr = self.head
        while count <= self.get_length():
            if itr not in container:
                container.add(itr)
            else:
                # self.remove_loop(itr)
                return True
            itr = itr.next
            count += 1
        return False
    
    def floyd_cycle(self):
        if self.head is None:
            print("Linked list is empty!")
            return False
        
        count = 0
        slow_p = fast_p = self.head
        while fast_p and fast_p.next:
            slow_p = slow_p.next
            fast_p = fast_p.next.next
            if slow_p == fast_p:
                # self.remove_loop(slow_p)
                return True
            count += 1
        return False
        
    def remove_loop(self, itr):
        p1 = p2 = itr
        k = 1
        
        while p1 != p2.next:
            k += 1
            p2 = p2.next
            
        p1 = p2 = self.head
        for i in range(k):
            p2 = p2.next
            
        while p1 != p2:
            p1 = p1.next
            p2 = p2.next
            
        while p1 != p2.next:
            p2 = p2.next
        p2.next = None
    
    def rotate_left(self, k):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        rotate = k%self.get_length()
        count = 0
        itr = self.head
        while count < rotate:
            self.delete_at_beginning()
            self.insert_at_end(itr.data)
            itr = itr.next
            count += 1
            
    def __delete_at_end(self):
        fnode = itr = self.head
        while itr.next != fnode:
            itr = itr.next
        return itr  
    
    def rotate_right(self, k):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        rotate = k%self.get_length()
        count = 0
        itr = self.head
        while count < rotate:
            val = self.__delete_at_end()
            self.delete_at_end()
            self.insert_at_beginning(val.data)
            itr = itr.next
            count += 1
            
    def get_first_node(self):
        if self.head is None:
            print("The Linked List is empty!")
            return
        return self.head
            
    def get_last_node(self):
        if self.head is None:
            print("The Linked list is empty!")
            return
        
        f_node = itr = self.head
        while itr.next != f_node:
            itr = itr.next
        return itr

if __name__ == "__main__":
    ll = LinkedList()
    ll.insert_at_beginning(10)
    ll.insert_at_beginning(20)
    ll.insert_at_beginning(30)
    ll.insert_at_end(40)
    ll.insert_at_end(90)
    ll.insert_at_end(50)
    ll.print_forward()
    ll.print_backward()
    ll.insert_values([110,35,26,3])
    ll.print_forward()
    ll.insert_at(0,1)
    ll.insert_at(13,100)
    print(ll.get_length())
    ll.insert_at(11, 100)
    ll.insert_at(5,12)
    ll.print_forward()
    print(ll.search(11))
    print(ll.search(100))
    print(ll.search(1))
    ll.insert_after_value(12, 15)
    ll.insert_after_value(100, 170)
    ll.insert_after_value(13, 111)
    ll.print_forward()
    ll.insert_before_value(12, 45)
    ll.insert_before_value(170, 10)
    ll.insert_before_value(1, 111)
    ll.print_forward()
    ll.delete_at(0)
    print(ll.get_length())
    ll.delete_at(17)
    ll.delete_at(16)
    ll.delete_at(6)
    ll.print_forward()
    ll.delete_at_end()
    ll.delete_at_beginning()
    ll.print_forward()
    ll.delete_by_value(30)
    ll.delete_by_value(100)
    ll.delete_by_value(170)
    ll.print_forward()
    ll.delete_after_value(15)
    ll.delete_after_value(20)
    ll.delete_after_value(3)
    ll.delete_after_value(10)
    ll.print_forward()
    ll.delete_before_value(15)
    ll.delete_before_value(20)
    ll.delete_before_value(40)
    ll.print_forward()
    ll.sort()
    ll.print_forward()
    ll.reverse()
    ll.print_forward()
    print(ll.detect_cycle())
    print(ll.floyd_cycle())
    print(ll.get_length())
    ll.rotate_left(5)
    ll.print_forward()
    print(ll.get_length())
    ll.rotate_right(3)
    ll.print_forward()
    # print(ll._LinkedList__get_last_node().data) (if function is mangled like __get_last_node(self))
    print(ll.get_first_node().data)
    print(ll.get_last_node().data)
```
