```python

#Singly Linked List:

class Node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
    
class LinkedList:
    def __init__(self):
        self.head = None
    
    def insert_at_beginning(self, data):
        self.head = Node(data, self.head)
        return 
    
    def insert_at_end(self, data):
        itr = self.head
        while itr.next:
            itr = itr.next
        itr.next = Node(data, itr.next)
        
    def delete_at_beginning(self):
        if self.head is None:
            print("Linked list is empty!")
            return 
        
        self.head = self.head.next
        
    def delete_at_end(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        if self.get_length() == 1:
            self.head = None
            return
        
        itr = self.head
        while itr.next.next:
            itr = itr.next
        itr.next = itr.next.next
    
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count += 1
            itr = itr.next
        return count
    
    def print_forward(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        itr = self.head
        fstr = ""
        while itr.next:
            fstr += str(itr.data) + "-->"
            itr = itr.next
        fstr += str(itr.data)
        print(fstr)
        
    def print_backward(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        itr = self.head
        bstr = ""
        while itr.next:
            bstr =  "<--" + str(itr.data) + bstr
            itr = itr.next
        bstr =  str(itr.data) + bstr
        print(bstr)
        
    def insert_values(self, data_list):
        for data in data_list:
            self.insert_at_end(data)
         
    def insert_at(self, index, data):
        if index < 0 or index > self.get_length():
            print("Invalid index!")
            return
        
        if index == 0:
            self.insert_at_beginning(data)
            return
        
        itr = self.head
        count = 0
        while itr:
            if count == index - 1:
                itr.next = Node(data, itr.next)
                return 
            count += 1
            itr = itr.next
    
    def delete_at(self, index):
        if self.head is None:
            print("Linked list is empty!")
            return False 
        
        if index < 0 or index >= self.get_length():
            print("Invalid Index")
            return
        
        if index == 0:
            self.delete_at_beginning()
            return
        
        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                itr.next = itr.next.next
                return
            count += 1
            itr = itr.next
            
    def search(self, val):
        if self.head is None:
            print("Linked list is empty!")
            return False
        
        itr = self.head
        while itr:
            if itr.data == val:
                return True
            itr = itr.next
        return False
    
    def insert_after_value(self, val, data):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        itr = self.head
        while itr:
            if itr.data == val:
                itr.next = Node(data, itr.next)
                return
            itr = itr.next
        
        print("No value found to insert after!")
    
    def insert_before_value(self, val, data):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        if self.head.data == val:
            self.insert_at_beginning(data)
            return
        
        itr = self.head
        while itr.next:
            if itr.next.data == val:
                itr.next = Node(data, itr.next)
                return
            itr = itr.next
        
        print("No value found to insert before!")
    
    def delete_by_value(self, val):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        if self.head.data == val:
            self.head = self.head.next
            return
        
        itr = self.head
        while itr.next:
            if itr.next.data == val:
                itr.next = itr.next.next
                return
            itr = itr.next
        
        print("No value found to delete!")
    
    def delete_after_value(self, val):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        itr = self.head
        while itr:
            if itr.data == val:
                if itr.next:
                    itr.next = itr.next.next
                    return
                else:
                    print(f"No node available to delete after value {val}")
                    return
            itr = itr.next
            
        print("No value found to delete after!")
    
    def delete_before_value(self, val):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        if self.head.data == val:
            print(f"No node to delete before the HEAD node {val}")
            return
        
        if self.head.next.data == val:
            self.head = self.head.next
            return
            
        itr = self.head
        while itr.next.next:
            if itr.next.next.data == val:
                itr.next = itr.next.next
                return
            itr = itr.next
                    
        print("No value found to delete before!")
    
    def sort(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        swapped = True
        while swapped:
            itr = self.head
            swapped = False
            while itr.next:
                if itr.data > itr.next.data:
                    itr.next.data, itr.data = itr.data, itr.next.data
                    swapped = True 
                itr = itr.next
                
            if not swapped:
                break
    
    def reverse(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        itr = self.head
        self.head = None
        while itr:
            self.insert_at_beginning(itr.data)
            itr = itr.next
    
    def detect_cycle(self):
        if self.head is None:
            print("Linked list is empty!")
            return False
            
        itr = self.head
        container = set()
        while itr:
            if itr not in container:
                container.add(itr)
            else:
                # self.remove_loop(itr)
                return True
            itr = itr.next
        return False   
    
    def rotate_left(self, k):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        rotate = k%self.get_length()
        if rotate == 0:
            return 
        else:
            count = 0
            itr = self.head
            while count != rotate:
                self.delete_at_beginning()
                self.insert_at_end(itr.data)
                count += 1
                itr = itr.next
                
    def __delete_at_end_node(self):
        itr = self.head
        while itr.next.next:
            itr = itr.next
        return itr.next
    
    def rotate_right(self, k):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        rotate = k%self.get_length()
        if rotate == 0:
            return 
        else:
            count = 0
            itr = self.head
            while count != rotate:
                node = self.__delete_at_end_node()
                self.delete_at_end()
                self.insert_at_beginning(node.data)
                count += 1
                itr = itr.next
    
    def floyd_cycle(self):
        if self.head is None:
            print("Linked list is empty!")
            return False
        
        slow_p = fast_p = self.head
        while fast_p and fast_p.next:
            slow_p = slow_p.next
            fast_p = fast_p.next.next
            if slow_p == fast_p:
                # self.remove_loop(slow_p)
                return True
        return False
    
    def remove_loop(self, itr):
        p1 = p2 = itr
        k = 1
        
        while p1 != p2.next:
            k += 1
            p2 = p2.next
            
        p1 = p2 = self.head
        for i in range(k):
            p2 = p2.next
            
        while p1 != p2:
            p1 = p1.next
            p2 = p2.next
        
        while p1 != p2.next:
            p2 = p2.next
        p2.next = None    
        
    def get_first_node(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        return self.head
            
    def get_last_node(self):
        if self.head is None:
            print("Linked list is empty!")
            return
        
        itr = self.head
        while itr.next:
            itr = itr.next
        return itr
            
if __name__ == "__main__":
    ll = LinkedList()
    ll.insert_at_beginning(10)
    ll.insert_at_beginning(20)
    ll.insert_at_beginning(30)
    ll.insert_at_end(40)
    ll.insert_at_end(90)
    ll.insert_at_end(50)
    ll.print_forward()
    ll.print_backward()
    ll.insert_values([110,35,26,3])
    ll.print_forward()
    ll.insert_at(0,1)
    ll.insert_at(13,100)
    print(ll.get_length())
    ll.insert_at(11, 100)
    ll.insert_at(5,12)
    ll.print_forward()
    print(ll.search(11))
    print(ll.search(100))
    print(ll.search(1))
    ll.insert_after_value(12, 15)
    ll.insert_after_value(100, 170)
    ll.insert_after_value(13, 111)
    ll.print_forward()
    ll.insert_before_value(12, 45)
    ll.insert_before_value(170, 10)
    ll.insert_before_value(1, 111)
    ll.print_forward()
    ll.delete_at(0)
    print(ll.get_length())
    ll.delete_at(17)
    ll.delete_at(6)
    ll.print_forward()
    ll.delete_at_end()
    ll.delete_at_beginning()
    ll.print_forward()
    ll.delete_by_value(30)
    ll.delete_by_value(100)
    ll.delete_by_value(170)
    ll.print_forward()
    ll.delete_after_value(15)
    ll.delete_after_value(20)
    ll.delete_after_value(10)
    ll.print_forward()
    ll.delete_before_value(15)
    ll.delete_before_value(20)
    ll.print_forward()
    ll.sort()
    ll.print_forward()
    ll.reverse()
    ll.print_forward()
    print(ll.detect_cycle())
    print(ll.floyd_cycle())
    print(ll.get_length())
    ll.rotate_left(5)
    ll.print_forward()
    print(ll.get_length())
    ll.rotate_right(3)
    ll.print_forward()
    print(ll.get_first_node().data)
    print(ll.get_last_node().data)
```
