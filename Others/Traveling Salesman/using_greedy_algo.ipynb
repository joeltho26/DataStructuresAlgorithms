```python
# travelling salesman - using greedy algorithm

class Graph:
    def __init__(self,size):
        self.size = size
        self.adj_matrix = [[0] * self.size for _ in range(self.size)]
        self.vertix_data = [""] * self.size
        
    def add_edge(self,u,v,weight):
        if 0 <= u < self.size and 0 <= v < self.size:
            self.adj_matrix[u][v] = weight
            self.adj_matrix[v][u] = weight
    
    def add_vertix_data(self,index,node):
        self.vertix_data[index] = node

def travelling_salesman(distances):
    size = len(distances)
    visited = [False] * size
    route = [0]
    visited[0] = True
    total_distance = 0
    for _ in range(1,size):
        last = route[-1]
        nearest = None
        min_dist = float('inf')
        for i in range(size):
            if not visited[i] and distances[last][i] < min_dist:
                min_dist = distances[last][i]
                nearest = i
            
        route.append(nearest) 
        visited[nearest] = True
        total_distance += min_dist
    
    total_distance += distances[route[-1]][0]
    route.append(0)
    return route, total_distance   

#since starting point is A else the nodes and the vertices are re-arranged
nodes = ["A","B","C","D","E","F"]
size = len(nodes)
vertices = [("A","B",2),("A","C",2),("A","D",5),("A","E",9),("A","F",3),
            ("B","A",2),("B","C",4),("B","D",6),("B","E",7),("B","F",8),
            ("C","A",2),("C","B",4),("C","D",8),("C","E",6),("C","F",3),
            ("D","A",5),("D","B",6),("D","C",8),("D","E",4),("D","F",9),
            ("E","A",9),("E","B",7),("E","C",6),("E","D",4),("E","F",10),
            ("F","A",3),("F","B",8),("F","C",3),("F","D",9),("F","E",10)]

g = Graph(size)
for idx,i in enumerate(nodes):
    g.add_vertix_data(idx,i)
    
for i in vertices:
    src = g.vertix_data.index(i[0])
    dest = g.vertix_data.index(i[1])
    g.add_edge(src,dest,i[2])
    
print(g.adj_matrix)
print(travelling_salesman(g.adj_matrix))
```
